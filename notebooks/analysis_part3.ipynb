# Indicators for Evaluating Company Strength

---

## 1. Profitability

- **Profit Margin (`profit_margins`)**
  - **What it measures:** The percentage of revenue that turns into net profit.
  - **Interpretation:** A higher and stable or growing margin suggests that the company efficiently converts sales into profit.

- **Return on Assets (ROA)**
  - **What it measures:** How efficiently the company uses its assets to generate profit.
  - **Interpretation:** A higher ROA indicates effective use of assets in generating earnings.

- **Return on Equity (ROE)**
  - **What it measures:** The profitability relative to shareholders’ equity.
  - **Interpretation:** A high and stable ROE implies the company is good at rewarding its shareholders.

---

## 2. Financial Stability (Leverage & Cash Flow)

- **Debt-to-Equity Ratio (`debt_to_equity`)**
  - **What it measures:** The level of debt compared to shareholders’ equity.
  - **Interpretation:** A very high ratio may signal over-leverage, making the company more vulnerable during downturns. Moderate or low values indicate a more stable balance sheet.

- **Free Cash Flow (`free_cashflow`)**
  - **What it measures:** The cash generated after covering operational and capital expenditures.
  - **Interpretation:** Consistently positive and growing free cash flow suggests the company can cover its debt, pay dividends, or reinvest in growth.

- **Payout Ratio (`payout_ratio`)**
  - **What it measures:** The percentage of net income paid out as dividends.
  - **Interpretation:** Extremely high payout ratios (above 80-90% or exceeding net income) may indicate the company has little room for reinvestment, which could be a sign of fragility.

---

## 3. Growth

- **Revenue Growth (`revenue_growth`)**
  - **What it measures:** The rate of change in revenue, typically year-over-year.
  - **Interpretation:** Consistent or accelerating revenue growth is a positive sign, suggesting market expansion or competitive advantage.

- **Earnings Growth (`earnings_growth`)**
  - **What it measures:** The rate at which net income is growing.
  - **Interpretation:** When combined with revenue growth, steady earnings growth shows the company’s ability to convert sales into profits over time.

- **Market Capitalization (`market_cap`)**
  - **What it measures:** The size of the company as perceived by the market.
  - **Interpretation:** Tracking changes in market cap (if you have historical data) can provide insights into investor sentiment regarding the company’s stability and growth prospects.

---

## 4. Market Valuation

- **Price/Earnings Ratios (`pe_trailing` and `pe_forward`)**
  - **What they measure:**
    - *Trailing P/E* uses historical earnings.
    - *Forward P/E* is based on forecasted earnings.
  - **Interpretation:**
    - Very high P/E ratios might indicate growth expectations but could also signal overvaluation.
    - Lower P/E ratios might suggest undervaluation, but they must be analyzed in the context of actual profitability and future prospects.

- **Price-to-Sales (`price_to_sales`)**
  - **What it measures:** How much the market is paying for each dollar of revenue.
  - **Interpretation:** Lower values may indicate a relatively cheaper valuation, though this metric is more relevant for companies with lower or negative earnings.

- **Price-to-Book (`price_to_book`)**
  - **What it measures:** The ratio of the current share price to its book value per share.
  - **Interpretation:** Particularly useful in sectors where the book value is a critical measure of underlying assets (e.g., finance or industrial sectors).

---

## 5. Stability & Liquidity

- **Beta**
  - **What it measures:** The volatility of the stock relative to the overall market.
  - **Interpretation:**
    - A beta near or below 1 indicates that the stock tends to move in line with or less than the market, which may suggest lower risk.
    - However, a low beta alone does not guarantee business stability; it might simply indicate lower trading activity or a defensive sector.

- **Average Volume (`average_volume`)**
  - **What it measures:** The average trading volume, indicating liquidity.
  - **Interpretation:** Higher trading volumes often correlate with established, well-known companies and greater liquidity, reducing trading impact.

- **Short Ratio (`short_ratio`)**
  - **What it measures:** The number of days needed to cover short positions at the average trading volume.
  - **Interpretation:**
    - A very high short ratio might indicate negative investor sentiment.
    - However, it can also sometimes create conditions for a short squeeze, which is an opportunity rather than a warning sign.

---

## 6. Qualitative and Contextual Factors

- **Sector and Industry (`sector`, `industry`)**
  - **Interpretation:** Even solid companies can show different performance characteristics depending on the cyclicality and dynamics of their sector. Defensive sectors (like utilities or consumer staples) are generally more stable compared to cyclical ones (like automotive or raw materials).

- **Country (`country`)**
  - **Interpretation:** The macroeconomic, regulatory, and political context of the country where the company is based can affect its risk profile.

- **Dividends (`dividend_yield`, `last_annual_dividend`)**
  - **Interpretation:** A stable or growing dividend, supported by sustainable earnings and cash flow, is often viewed as a sign of financial strength and confidence in future performance.

---

## Summary

- **Profitability & Growth:**  
  Look for companies with stable or increasing profit margins, ROA, and ROE, along with steady revenue and earnings growth.

- **Financial Stability:**  
  Favor firms with a moderate or low debt-to-equity ratio, robust and growing free cash flow, and a sensible payout ratio that balances reward with reinvestment.

- **Market Valuation:**  
  Evaluate the P/E ratios (both trailing and forward), price-to-sales, and price-to-book ratios in the context of the sector to assess whether the stock is fairly valued.

- **Stability & Liquidity:**  
  Consider a lower beta, high average volume, and a reasonable short ratio as indicators of less volatile and more liquid stocks.

- **Contextual Factors:**  
  Factor in the company’s sector, industry, country of operation, and dividend policies to round out the assessment.

These combined indicators offer a comprehensive framework to differentiate companies with solid, resilient business models from those that may be more financially fragile.

* **`plt.tight_layout(rect=[0, 0.03, 1, 0.95])`**

  This function automatically adjusts the spacing between subplots so that all elements (like axes labels, tick labels, and titles) fit within the figure area without overlapping. The `rect` parameter specifies a rectangle in normalized figure coordinates (where 0 is the start and 1 is the end) that the subplots should fit into. In this case, `rect=[0, 0.03, 1, 0.95]` means:
  
  - **Left boundary** is at `0` (the very left edge of the figure).
  - **Bottom boundary** is at `0.03` (leaving a 3% margin at the bottom).
  - **Right boundary** is at `1` (the very right edge of the figure).
  - **Top boundary** is at `0.95` (leaving a 5% margin at the top).

* **`plt.show()`**

  This function displays the final figure after all adjustments have been made.

```python
# Convert the 'settore' column to a categorical data type.
# Converting a column to 'category' optimizes memory usage and provides access to additional methods
# for handling categorical data, such as .cat.codes and .cat.categories.
df['sector'] = df['sector'].astype('category')

# .cat.codes returns a Series of integer codes corresponding to each category in the 'settore' column.
# Each unique category is mapped to a unique integer value. This is useful for plotting because
# histograms require numerical input.
sector_codes = df['sector'].cat.codes

# .cat.categories returns an Index of the original category labels for the 'settore' column.
# These labels can be used to replace the numeric ticks on the x-axis with the actual category names.
sector_labels = df['sector'].cat.categories

# Create a new figure with a specified size.
plt.figure(figsize=(10, 6))

# Plot the histogram using plt.hist().
# - 'sector_codes' provides the numerical representations of the categories.
# - 'bins' is set to the number of unique categories to ensure each category gets its own bin.
# - 'rwidth=0.8' sets the width of the bars to 80% of the bin width, adding padding between bars.
# - 'color' is set to light blue, and no edgecolor is specified.
plt.hist(sector_codes, bins=len(sector_labels), rwidth=0.8, color='lightblue')

# Set the title of the plot.
plt.title('Distribution of Settore Values')

# Label the x-axis and y-axis.
plt.xlabel('sector')
plt.ylabel('Frequency')

# Replace the numeric x-axis ticks with the actual category labels.
# The 'ticks' are a range from 0 to the number of unique categories.
# The 'labels' are the corresponding original sector names.
# The labels are rotated by 90 degrees for improved readability.
plt.xticks(ticks=range(len(sector_labels)), labels=sector_labels, rotation=90)

# Adjust the layout to ensure that all plot elements fit neatly in the figure area.
plt.tight_layout()

# Display the histogram.
plt.show()
```

```python
# Extract the top 25 companies with the largest market capitalization.

# df.nlargest(n, columns) returns a subset with the top 'n' rows based on
# the specified column(s). Here, we use "market_cap" to determine the size.
top25 = df.nlargest(25, 'market_cap')

# Sort these 25 companies in descending order of their market capitalization.
# The first rows represent companies with the highest market cap.
top25_sorted = top25.sort_values('market_cap', ascending=False)
```

```python
# Define the list of columns to plot. We removed 'pe_trailing' and 'pe_forward' from the list,
# so now we only include 'market_cap', 'beta', 'net_income', and 'profit_margins'.
columns_to_plot = ['market_cap', 'beta']

# Iterate over each column name in the list to create an individual plot for each metric.
for column in columns_to_plot:

    # Create a new figure for the current column. This ensures that each plot is separate.
    plt.figure()

    # Assume that 'top25' is a pre-existing pandas DataFrame containing data for the top 25 companies.
    # For the current column, remove any missing (NaN) values using dropna() to ensure only valid numbers are plotted.
    data = top25[column].dropna()

    # Calculate key statistical values for the data in the current column:
    # q1_val: The 25th percentile (first quartile).
    q1_val = data.quantile(0.25)
    # q3_val: The 75th percentile (third quartile).
    q3_val = data.quantile(0.75)
    # median_val: The median (50th percentile) value.
    median_val = data.median()
    # iqr_val: The interquartile range (IQR) is the difference between Q3 and Q1.
    iqr_val = q3_val - q1_val
    # lower_whisker: The lower boundary for detecting outliers, calculated as Q1 minus 1.5 times the IQR.
    lower_whisker = q1_val - 1.5 * iqr_val
    # upper_whisker: The upper boundary for detecting outliers, calculated as Q3 plus 1.5 times the IQR.
    upper_whisker = q3_val + 1.5 * iqr_val

    # Create a boxplot for the data.
    # - 'whis=1.5' specifies the factor used to calculate the whiskers (1.5 * IQR).
    # - 'showfliers=True' ensures that outlier markers (fliers) are displayed on the plot.
    # We are using the default styling for the boxplot elements.
    plt.boxplot(
        data,
        whis=1.5,
        showfliers=True
    )

    # Add numeric annotations for key statistical values.
    # The default boxplot positions the box at x=1. We will annotate slightly to the right of the box (x=1.1).
    # The annotations include:
    #   Q1 (first quartile)
    #   Q2 (median)
    #   Q3 (third quartile)
    #   Lower whisker (Q1 - 1.5*IQR)
    #   Upper whisker (Q3 + 1.5*IQR)

    # Annotate Q1: Place the text to the right of the box plot.
    plt.text(1.1, q1_val, f"Q1: {q1_val:.2f}", verticalalignment='center', fontsize=10)
    # Annotate the median (Q2)
    plt.text(1.1, median_val, f"Median: {median_val:.2f}", verticalalignment='center', fontsize=10)
    # Annotate Q3
    plt.text(1.1, q3_val, f"Q3: {q3_val:.2f}", verticalalignment='center', fontsize=10)

    # Set the title for the plot to indicate which column is being visualized.
    plt.title(f'Boxplot of {column} (with outliers shown)')

    # Label the y-axis with the current column name.
    plt.ylabel(column)

    # Enable grid lines for better readability.
    plt.grid(True)

    # Display the current plot.
    plt.show()
```

```python
# Loop over each column in the list to create an individual horizontal boxplot for each metric.
for column in columns_to_plot:

    # Create a new figure for the current column.
    plt.figure()

    # Assume that 'top25' is a pre-existing pandas DataFrame containing data for the top 25 companies.
    # For the current column, drop any missing values (NaN) to ensure only valid numeric data is used.
    data = top25[column].dropna()

    # Calculate key statistical values for the data:
    # q1_val: 25th percentile (first quartile)
    q1_val = data.quantile(0.25)
    # q3_val: 75th percentile (third quartile)
    q3_val = data.quantile(0.75)
    # median_val: 50th percentile (median)
    median_val = data.median()
    # iqr_val: The interquartile range (IQR) is the difference between Q3 and Q1.
    iqr_val = q3_val - q1_val
    # lower_whisker: The lower boundary for detecting outliers, calculated as Q1 minus 1.5 times the IQR.
    lower_whisker = q1_val - 1.5 * iqr_val
    # upper_whisker: The upper boundary for detecting outliers, calculated as Q3 plus 1.5 times the IQR.
    upper_whisker = q3_val + 1.5 * iqr_val

    # Create a horizontal boxplot for the data.
    # - whis=1.5 sets the multiplier for the IQR to determine the whisker length.
    # - showfliers=True ensures that outlier points are displayed.
    # - vert=False creates a horizontal boxplot.
    plt.boxplot(
        data,
        whis=1.5,
        showfliers=True,
        vert=False
    )

    # Identify outliers as values that fall below the lower whisker or above the upper whisker.
    outliers = data[(data < lower_whisker) | (data > upper_whisker)]

    # Annotate each outlier with its corresponding ticker symbol.
    # We assume that the DataFrame 'top25' contains a column 'symbol' that holds the ticker for each row.
    for idx, value in outliers.items():
        # Retrieve the ticker symbol from top25['symbol'] corresponding to the current index.
        ticker_label = top25.loc[idx, 'symbol']
        # Annotate the outlier on the horizontal boxplot.
        # The annotation is placed slightly below the boxplot (y=0.9) to avoid overlapping with the box.
        plt.text(value, 0.9, f"{ticker_label}", horizontalalignment='center', fontsize=10, color='red')

    # Create dummy line objects to include Q1, Median (Q2), and Q3 values in the legend.
    # These dummy handles will not be plotted but will serve to display the numeric values in the legend.
    dummy_q1 = mlines.Line2D([], [], color='none', label=f"Q1: {q1_val:.2f}")
    dummy_median = mlines.Line2D([], [], color='none', label=f"Median: {median_val:.2f}")
    dummy_q3 = mlines.Line2D([], [], color='none', label=f"Q3: {q3_val:.2f}")

    # Add the legend to the plot using the dummy handles.
    # The legend will display the Q1, Median, and Q3 values.
    plt.legend(handles=[dummy_q1, dummy_median, dummy_q3], loc='upper right', frameon=True)

    # Set the title of the plot, indicating which metric is being visualized.
    plt.title(f'Horizontal Boxplot of {column} (outliers labeled with ticker)')

    # Label the x-axis with the current column name.
    plt.xlabel(column)

    # Remove the y-axis tick marks since the boxplot represents a single category.
    plt.yticks([])

    # Enable grid lines on the plot to improve readability.
    plt.grid(True)

    # Display the plot.
    plt.show()
```

```python
# -------------------------------------------------------------------------
# 1) Define a Diverging Color Logic Based on "beta"
# -------------------------------------------------------------------------
# We want to create a diverging color scheme that highlights the range of "beta" values.
# Instead of the previous scheme (Orange -> Gray -> Green), we use:
#   - Orange for low beta values,
#   - Light Gray as the midpoint,
#   - Dark Blue for high beta values.
# This diverging colormap helps viewers quickly assess where each company's beta lies along the continuum.
colors_list = [
    (1.0, 0.65, 0.0),  # Orange for the lower bound (low beta)
    (0.8, 0.8, 0.8),   # Light Gray for the midpoint
    (0.0, 0.0, 0.5)    # Dark Blue for the upper bound (high beta)
]

# Create a linear segmented colormap that smoothly transitions through the three anchor colors.
# 'N=256' sets the resolution of the colormap to 256 discrete color levels.
cmap = mcolors.LinearSegmentedColormap.from_list(
    'OrangeLightGrayDarkBlue',
    colors_list,
    N=256
)

# -------------------------------------------------------------------------
# 2) Normalize "beta" Values
# -------------------------------------------------------------------------
# To map each beta value to a color, we first need to normalize the beta values to the [0, 1] interval.
# This normalization scales the smallest beta to 0 and the largest beta to 1.
min_beta = top25_sorted['beta'].min()
max_beta = top25_sorted['beta'].max()

# plt.Normalize creates a normalization function that will convert beta values to a 0-1 range.
norm = plt.Normalize(vmin=min_beta, vmax=max_beta)

# -------------------------------------------------------------------------
# 3) Assign Colors to Bars Based on Beta
# -------------------------------------------------------------------------
# We iterate over the "beta" values of the top25_sorted DataFrame.
# For each beta value:
#   1. Normalize the beta value to the [0, 1] range using norm(b).
#   2. Pass the normalized value to the colormap (cmap) to get the corresponding RGBA color.
# The resulting list 'bar_colors' will be used to color each bar in the chart.
bar_colors = []
for b in top25_sorted['beta']:
    # Normalize the beta value and retrieve the corresponding color from the colormap.
    color = cmap(norm(b))
    bar_colors.append(color)

# -------------------------------------------------------------------------
# 4) Plot Setup - Creating a Horizontal Bar Chart
# -------------------------------------------------------------------------
# Create a horizontal bar chart to visualize the companies.
# - y-axis: Company names (from "company_name" column).
# - Bar width: Percentage of total market cap (from "pct_of_total" column).
# We use a horizontal bar chart (barh) because it better accommodates long company names.
plt.figure(figsize=(12, 8))
bars = plt.barh(
    y=top25_sorted["company_name"],
    width=top25_sorted["pct_of_total"],
    height=0.65,
    color=bar_colors  # Each bar is colored based on its beta value
)

# -------------------------------------------------------------------------
# 5) Adding Labels and Title
# -------------------------------------------------------------------------
# Set axis labels and title to clearly describe the plot.
plt.xlabel("Percentage of Total Market Cap (%)")
plt.ylabel("Company Name")
plt.title("Top 25 Companies by Market Cap (% of Total) - Colored by Beta")

# Remove the top and right spines (borders) of the plot for a cleaner appearance.
ax = plt.gca()  # Get the current Axes instance
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# -------------------------------------------------------------------------
# 6) Label Each Bar with Its Percentage Value
# -------------------------------------------------------------------------
# Iterate over the bar objects to annotate each bar with its exact percentage value.
# - width: The length of the bar (percentage value).
# - y_pos: The vertical center of the bar.
# We place the text slightly to the right of each bar.
for bar in bars:
    width = bar.get_width()  # Retrieve the width (percentage) of the bar
    y_pos = bar.get_y() + bar.get_height() / 2  # Calculate the vertical midpoint of the bar
    plt.text(width + 0.1, y_pos, f"{width:.2f}%", va="center")  # Annotate the bar with the percentage

# -------------------------------------------------------------------------
# 7) (Optional) Add a Colorbar Showing How Beta Maps to the Colormap
# -------------------------------------------------------------------------
# Create a ScalarMappable object to connect the colormap and normalization.
# This object is used to generate a colorbar that visualizes the mapping from beta values to colors.
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])  # Required for compatibility with older Matplotlib versions
cbar = plt.colorbar(sm, ax=ax)
# Label the colorbar to indicate that it represents the beta values.
cbar.set_label("Beta")

# -------------------------------------------------------------------------
# 8) Final Chart Rendering
# -------------------------------------------------------------------------
# Adjust layout parameters to ensure the plot elements fit well within the figure area.
plt.tight_layout()

# Display the final chart.
plt.show()
```

```python
import yfinance as yf                # Import the yfinance library to download stock data
import matplotlib.pyplot as plt      # Import matplotlib for plotting graphs

# List of tickers of interest: Apple, Amazon, Meta, Microsoft, and S&P500
tickers = ['AAPL', 'AMZN', 'META', 'MSFT', '^GSPC']

# Download stock data for the last 30 days using daily intervals.
# Note: To specify an exact date range, you can use the 'start' and 'end' parameters.
# Example: start='2023-02-01', end='2023-03-01'
data = yf.download(tickers, period='30d', interval='1d')

# Extract the closing prices from the downloaded data.
# Although the comment mentions "Adj Close", the code actually selects the "Close" column.
closing_prices = data['Close']

# Create a new figure for plotting with a specified size (width: 10, height: 6 inches)
plt.figure(figsize=(10, 6))

# Loop through each ticker and plot its closing prices against the dates.
for ticker in tickers:
    plt.plot(closing_prices.index, closing_prices[ticker], label=ticker)

# Set the title and labels for the plot to provide context
plt.title("Comparison of closing prices over the last 30 days (including S&P500)")
plt.xlabel("Date")
plt.ylabel("Closing Price")

# Add a legend to differentiate between the tickers in the plot
plt.legend()

# Display the plot on the screen
plt.show()
```

```python

# Normalize the prices by dividing each day's price by the price on the first day.
# closing_prices.iloc[0] retrieves the first row of the closing_prices DataFrame,
# which corresponds to the closing prices on the first day of the period.
normalized_prices = closing_prices / closing_prices.iloc[0]

# Create a new figure for the plot with a specified size (10 inches wide, 6 inches tall)
plt.figure(figsize=(10, 6))

# Loop through each ticker and plot its normalized closing prices over time
for ticker in tickers:
    plt.plot(normalized_prices.index, normalized_prices[ticker], label=ticker)

# Set the title and labels for the plot to provide context
plt.title("Comparison of normalized closing prices over the last 30 days")
plt.xlabel("Date")
plt.ylabel("Normalized Price (relative to the first day)")

# Add a legend to differentiate between the tickers
plt.legend()

# Display the plot
plt.show()
```

```python
# Create a new figure for the plot with a specified size (10 inches wide, 6 inches tall)
plt.figure(figsize=(10, 6))

# Retrieve the final normalized value for S&P500 for later comparison
# .iloc[-1] retrieves the last row of the DataFrame, which corresponds to the final normalized price in the period.
sp500_final = normalized_prices['^GSPC'].iloc[-1]

# Loop through each ticker and plot its normalized closing prices over time
for ticker in tickers:
    if ticker == '^GSPC':
        # Plot S&P500 as a black, dashed line
        plt.plot(normalized_prices.index, normalized_prices[ticker], label=ticker, color='black', linestyle='--')
    else:
        # Retrieve the final normalized value for the current ticker
        # Here .iloc[-1] similarly retrieves the last available normalized price for the ticker.
        ticker_final = normalized_prices[ticker].iloc[-1]
        # If the ticker's final value is greater than S&P500's, assign light blue; otherwise, assign light grey
        color = 'lightblue' if ticker_final > sp500_final else 'lightgrey'
        plt.plot(normalized_prices.index, normalized_prices[ticker], label=ticker, color=color)

# Set the title and labels for the plot to provide context
plt.title("Comparison of normalized closing prices over the last 30 days")
plt.xlabel("Date")
plt.ylabel("Normalized Price (relative to the first day)")

# Add a legend to differentiate between the tickers
plt.legend()

# Display the plot
plt.show()
```

```python
# Create a new figure with a specified size (10 inches wide, 6 inches tall)
plt.figure(figsize=(10, 6))

# Retrieve the final normalized value for S&P500 for later comparison.
# .iloc[-1] retrieves the last row of the DataFrame, corresponding to the final normalized price.
sp500_final = normalized_prices['^GSPC'].iloc[-1]

# Get current axes to later adjust spines and add annotations
ax = plt.gca()

# Loop through each ticker to plot its normalized closing prices over time
for ticker in tickers:
    # Retrieve the final normalized value for the current ticker using .iloc[-1]
    ticker_final = normalized_prices[ticker].iloc[-1]

    # Set line style and color: S&P500 gets a black dashed line; others are colored based on comparison with S&P500.
    if ticker == '^GSPC':
        line_color = 'black'
        line_style = '--'
    else:
        line_color = 'lightblue' if ticker_final > sp500_final else 'lightgrey'
        line_style = '-'  # Solid line for non-S&P500 tickers

    # Plot the normalized prices line for the ticker with the selected style and color.
    plt.plot(normalized_prices.index, normalized_prices[ticker], color=line_color, linestyle=line_style)

    # Retrieve the final date and value for the current ticker.
    final_date = normalized_prices.index[-1]
    final_value = ticker_final

    # Overlay the final data point with a circle marker.
    # The marker uses the same color as the line and a black edge for better visibility.
    plt.scatter(final_date, final_value, color=line_color, edgecolor='black', zorder=5, s=100)

    # Calculate the variation relative to the first element.
    # Since the series is normalized, the first value is 1; therefore, variation = ticker_final - 1.
    # Multiply by 100 to convert the variation into a percentage.
    variation = (ticker_final - 1) * 100

    # Format the variation text with a plus or minus sign and percentage symbol.
    variation_text = f"({variation:+.2f}%)"

    # Annotate next to the marker:
    # xy specifies the marker's location, and xytext sets the offset (e.g., 15 points to the right).
    # textcoords='offset points' means the offset is measured in points.
    ax.annotate(f"{ticker} {variation_text}",
                xy=(final_date, final_value),
                xytext=(15, 0),           # Increase the horizontal offset to separate the text from the marker
                textcoords='offset points',
                verticalalignment='center',
                fontsize=10,
                color=line_color)

# Remove y-axis and frame; keep only the x-axis visible.
ax.spines['top'].set_visible(False)         # Remove top frame line
ax.spines['right'].set_visible(False)       # Remove right frame line
ax.spines['left'].set_visible(False)        # Remove left frame line
ax.yaxis.set_ticks([])                      # Remove y-axis tick marks
plt.ylabel("")                              # Remove y-axis label

# Set the title and x-axis label for context.
plt.title("Comparison of normalized closing prices over the last 30 days")
plt.xlabel("Date")

# Display the plot.
plt.show()
```

```python
# Define the ticker symbol for Microsoft (MSFT)
ticker = 'MSFT'

# Download historical data for the specified ticker for the last 30 days with daily intervals.
# The returned DataFrame includes several columns such as Open, High, Low, Close, Adj Close, and Volume.
data = yf.download(ticker, period='30d', interval='1d')

# Select the 'Close' column from the DataFrame, which represents the daily closing prices,
# and convert the resulting pandas Series into a NumPy array for further manipulation.
x = data['Close'].values

# Create a new figure and a single subplot (axis) with a specified size.
# The figsize parameter is set to (10, 1), meaning the plot will be 10 inches wide and 1 inch tall.
fig, ax = plt.subplots(1, 1, figsize=(10, 1))

# Plot the time series data as a continuous black line.
# np.arange(len(x)) creates an array of indices corresponding to each data point in x,
# which is used as the x-axis.
ax.plot(np.arange(len(x)), x, color='k')

# Plot a red circle marker on the final data point of the series.
# 'len(x) - 1' specifies the x-position of the last data point,
# and 'x[-1]' retrieves the last closing price.
ax.plot(len(x) - 1, x[-1], color='r', marker='o')

# Remove all plot borders (spines) for a cleaner appearance.
# The ax.spines dictionary contains the borders of the plot (top, right, bottom, left).
for _, spine in ax.spines.items():
    spine.set_visible(False)

# Remove tick marks and labels from both the x-axis and y-axis to simplify the visualization.
ax.set_xticks([])
ax.set_yticks([])

# Display the final plot.
plt.show()
```

```python
import yfinance as yf               # Import yfinance to download historical financial data from Yahoo Finance
import matplotlib.pyplot as plt     # Import matplotlib's pyplot module to create and customize plots
import numpy as np                  # Import NumPy for efficient numerical operations and array handling

# Define the ticker symbol for Microsoft (MSFT)
ticker = 'MSFT'

# Download historical data for the specified ticker over the last 30 days using a daily interval.
# The data returned includes several columns such as Open, High, Low, Close, Adj Close, and Volume.
data = yf.download(ticker, period='30d', interval='1d')

# Select the 'Close' column from the downloaded DataFrame, which represents the closing price for each day.
# Convert the resulting pandas Series into a NumPy array for easier manipulation in subsequent calculations.
x = data['Close'].values

# Create a new figure and a single subplot (axis) for plotting.
# The figsize parameter is set to (10, 1) which means the plot will be 10 inches wide and 1 inch tall,
# creating a slim, horizontal plot that emphasizes the time series line.
fig, ax = plt.subplots(1, 1, figsize=(10, 1))

# Plot the time series data as a continuous black line.
# np.arange(len(x)) generates an array of indices [0, 1, 2, ..., len(x)-1] to be used as the x-axis values,
# effectively mapping each closing price to its corresponding time step.
ax.plot(np.arange(len(x)), x, color='k')

# Overlay a red circle marker on the last data point of the series.
# The x-coordinate is set to len(x)-1 (the index of the last element) and the y-coordinate is x[-1] (the last closing price).
# This marker highlights the final closing price in the time series.
ax.plot(len(x) - 1, x[-1], color='lightgrey', marker='o')

# Annotate the last data point with the ticker symbol.
# The annotation is placed slightly to the right of the marker (offset of 15 points horizontally)
# and centered vertically relative to the data point.
ax.annotate(ticker,
            xy=(len(x) - 1, x[-1]),
            xytext=(15, 0),          # Offset the text 15 points to the right of the marker
            textcoords='offset points',
            verticalalignment='center',
            fontsize=10,
            color='k')


# Add a light grey transparent box between y=408 and y=423.
# ax.axhspan() creates a horizontal span (box) across the axis between the specified y-values.
# The color is set to light grey and alpha=0.5 makes the box semi-transparent.
ax.axhspan(408, 423, color='darkgrey', alpha=0.5)

# Add additional horizontal reference lines for further context.
# The next two horizontal lines are drawn in blue with a dash-dot linestyle.
ax.axhline(y=468, color='darkgrey', linestyle='-.')
ax.axhline(y=376, color='darkgrey', linestyle='-.')

# Add text labels for the 52 Week High and 52 Week Low lines.
# The text is positioned near the left edge (x=0) and aligned vertically to the respective y-value.
ax.text(0, 468, " 52 Weeks High", verticalalignment='bottom', color='black', fontsize=10)
ax.text(0, 376, " 52 Weeks Low", verticalalignment='bottom', color='black', fontsize=10)

# Remove all the plot borders (spines) for a cleaner and minimalist visual appearance.
# The ax.spines dictionary contains the top, right, bottom, and left borders of the plot.
for _, spine in ax.spines.items():
    spine.set_visible(False)

# Remove all tick marks and labels from both the x-axis and y-axis.
# This step is taken to further simplify the appearance of the plot, focusing attention solely on the line and markers.
ax.set_xticks([])
ax.set_yticks([])

# Finally, display the completed plot on the screen.
plt.show()
```
