```python
#Load data from output.csv file, randomly sampling 25 rows, and selecting the desired columns.

# Load the CSV file
import pandas as pd
df = pd.read_csv('output.csv')

# Randomly sample 25 elements from the DataFrame (using random_state for reproducibility)
df_sample = df.sample(50, random_state=11)

# Select only the desired columns: symbol, market_cap, and sector
df_sample = df_sample[['symbol', 'market_cap', 'sector']]

# Display the initial sampled DataFrame
print('--- Initial Sampled DataFrame ---')
print(df_sample)
```

```python
# Display general information about the DataFrame
print('--- DataFrame Info ---')
df_sample.info()
print('\n--- DataFrame Description ---')
print(df_sample.describe())
print('\n--- DataFrame Description Objects ---')
print(df_sample.describe(include='object'))
```

```python
#Show the content by position and by column

# Show the first element
print('--- First Element ---')
print(df_sample.iloc[0])

# Show the market_cap value of the first element
print('\n--- Market Cap of the first Element ---')
print(df_sample.iloc[0]['market_cap'])

# Show the last element
print('\n--- Last Element ---')
print(df_sample.iloc[-1])

# Show the second-to-last element
print('\n--- Second-to-Last Element ---')
print(df_sample.iloc[-2])
```

```python
# Sort the DataFrame by Market_cap
df_sorted = df_sample.sort_values(by='market_cap', ascending=False)
print('--- DataFrame Sorted by Market_cap ---')
print(df_sorted)
```

## Grouping and Aggregation

In this section, we group the DataFrame by the `sector` column and apply multiple aggregation functions on the `market_cap` column. We use:

- **sum**: to get the total market cap per sector,
- **mean**: to calculate the average market cap per sector,
- **count**: to know how many entries each sector has.

The result is reset into a new DataFrame.

```python
# Group the DataFrame by 'sector' and aggregate 'market_cap' with multiple functions
df_grouped = df_sample.groupby('sector')['market_cap'].agg(['sum', 'mean', 'count']).reset_index()

# Display the aggregated DataFrame by sector
print('--- Aggregated DataFrame by Sector ---')
print(df_grouped)

# Group the DataFrame by 'sector' and aggregate 'market_cap' with multiple functions
df_grouped = df_sample.groupby('sector')['market_cap'].agg(['sum', 'mean', 'count', 'min', 'max']).reset_index()

# Sort the aggregated DataFrame by the 'count' column in descending order
df_grouped = df_grouped.sort_values(by='count', ascending=False)

# Display the aggregated DataFrame by sector
print('--- Aggregated DataFrame by Sector ---')
print(df_grouped)
```

##Advanced sorting and Indexing: Using `set_index()`, Multi-Indexing, and Sorting by Multiple Columns

This section demonstrates advanced indexing and sorting techniques:

- **Setting the index**: We use `set_index()` to set the `symbol` column as the index.
- **Multi-indexing**: We create a multi-index DataFrame by setting both `sector` and `symbol` as the index.
- **Sorting**: We sort the DataFrame by `sector` (ascending) and then by `market_cap` (descending within each sector) using `sort_values()`.

```python
# Set 'symbol' as the index using set_index()
df_indexed = df_sample.set_index('symbol')
print('--- DataFrame with symbol as index ---')
print(df_indexed)

# Create a multi-index DataFrame using both 'sector' and 'symbol'
df_multi_indexed = df_sample.set_index(['sector', 'symbol'])
print('\n--- DataFrame with Multi-Index (sector, symbol) ---')
print(df_multi_indexed)

# Sort the DataFrame by 'sector' (ascending) and then by 'market_cap' (descending)
df_sorted_multi = df_sample.sort_values(by=['sector', 'market_cap'], ascending=[True, False])
print('\n--- DataFrame sorted by sector and market_cap (descending within each sector) ---')
print(df_sorted_multi)

```

## Filtering and Conditional Selection with Boolean Masks

Here we filter the DataFrame using cthe following conditions:

- First, we compute the average of the `market_cap` column.
- Then, we apply a filter to keep only the rows where the `market_cap` is greater than this average.
- Additionally, we filter to include only rows where the `sector` is either "Technology" or "Financial Services" using the `isin()` method.

This is accomplished by creating a boolean mask that combines the conditions using the logical AND operator (`&`).

```python
# Calculate the average market_cap value
average_market_cap = df_sample['market_cap'].mean()

# Create a boolean mask for filtering
condition = (df_sample['market_cap'] > average_market_cap) & (df_sample['sector'].isin(['Technology', 'Financial Services']))

# Apply the filter to the DataFrame
df_filtered = df_sample[condition]

# Display the filtered DataFrame
print('--- Filtered DataFrame with market_cap > average and specific sectors ---')
print(df_filtered)

#modify the condition in creatiing a condition where we take the company with a market cap greater than the average and in the industrials sector

condition = (df_sample['market_cap'] > average_market_cap) & (df_sample['sector'] == 'Industrials')
# Apply the filter to the DataFrame
df_filtered = df_sample[condition]
# Display the filtered DataFrame
print('--- Filtered DataFrame with market_cap > average and in the Industrials sector ---')
print(df_filtered)
# Save the filtered DataFrame to a new CSV file
#df_filtered.to_csv('filtered_output.csv', index=False)
# Save the DataFrame to a CSV file
#df_sample.to_csv('output.csv', index=False)
```

```python
#Operations on non-numerical data
# Get unique values in the 'sector' column
unique_sectors = df_sample['sector'].unique()
print('--- Unique Sectors ---')
print(unique_sectors)

# Count the occurrences of each unique value in the 'sector' column
sector_counts = df_sample['sector'].value_counts()
print('\n--- Count of Each Sector ---')
print(sector_counts)

# Convert the 'sector' column to lowercase
df_sample['sector_upper'] = df_sample['sector'].str.upper()
print('\n--- First version ---')
print(df_sample)

# Filter rows where the 'sector' column contains the substring "tech"
# (case-insensitive search; na=False avoids errors with NaN values)
df_tech = df_sample[df_sample['sector'].str.contains('tech', case=False, na=False)]
print('\n--- Tech version ---')
print(df_tech)

# Replace specific values in the 'sector' column
# For example, replace "Financial Services" with "Finance"
df_sample['sector_replaced'] = df_sample['sector'].replace({'Financial Services': 'Finance'})
print('\n--- One more version ---')
print(df_sample)

# Map the sectors to a simplified version using a dictionary
sector_mapping = {
    'Technology': 'Tech',
    'Financial Services': 'Finance',
    'Industrials': 'Industrial'
}
df_sample['sector_mapped'] = df_sample['sector'].map(sector_mapping)
print('\n--- One more again ---')
print(df_sample)
```
